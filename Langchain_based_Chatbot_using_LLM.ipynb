```python
import os,pickle
from langchain.document_loaders import TextLoader
from langchain.text_splitter import CharacterTextSplitter
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import FAISS
from langchain_google_genai import GoogleGenerativeAI, ChatGoogleGenerativeAI
from langchain.chains import RetrievalQA
```

```python
os.environ["GOOGLE_API_KEY"]="AIzaSyA5WmRhYs-vlLOFxhQPDdfh9ndIVirGV0Q"
```

```python
loader=TextLoader("para1.txt")
docs_raw=loader.load()
splitter=CharacterTextSplitter(chunk_size=500,chunk_overlap=50)
docs=splitter.split_documents(docs_raw)
```

```python
embeddings=HuggingFaceEmbeddings()
vectorstore=FAISS.from_documents(docs,embeddings)
```

```python
vectorstore.save_local("faiss_index")
```

```python
llm=ChatGoogleGenerativeAI(model="models/gemini-1.5-flash", temperature=0.5)
qa_chain=RetrievalQA.from_chain_type(llm=llm,retriever=vectorstore.as_retriever())
```

```python
print(qa_chain.run("What is Bali?"))
```
